```python
# 使用pandas的 read_csv()讀入檔案，預設參數是 pd.read_csv，之中會有幾個參數要注意
# sep = 預設為逗點
# header = None，表示檔案中沒有變數名稱(欄位)，預設檔案讀取的第一列都是變數header = YES
# names  = 如果有指定header = None，就需要使用names來給變數名稱
# skiprows = 要略過多少列檔案上方資料
# skipfooder = 要略過多少列檔案下方資料
# nrows = 指定要讀入多少列資料
# na_values = 除了內建的遺漏值之外，還有哪些額外的字元需要成為遺漏值 
import pandas as pd
csv_url1 = "http://storage.googleapis.com/ds_data_import/chicago_bulls_1995_1996.csv"
csv_df1 = pd.read_csv(csv_url1, header = None, skiprows = 1, names = ['number' , 'player', 'pos', 'ht', 'wt', 'birth_date', 'college']) 
csv_df1
```

```python
# 使用pandas的read_table()讀入txt檔案, 設定參數為pd.read_table
# 預設 sep = /t 意即 tab鍵
import pandas as pd
txt_url = "http://storage.googleapis.com/ds_data_import/chicago_bulls_1995_1996.txt"
txt_df = pd.read_table(txt_url, sep = ";")
txt_df
```

```python
# 使用 pandas 的 pd.read_excel()來讀取excel檔案
# sheet_name = 預設為排序最前面的工作表，可以使用整數選擇，也可以使用工作表的名稱來選擇
# header = 預設為以最上方的一列作為變數名稱
# usecols = 選擇哪幾個變數要載入
# skiprows & skipfooter 皆和read_csv的參數相同
import pandas as pd
xlsx_url = "http://storage.googleapis.com/ds_data_import/fav_nba_teams.xlsx"
chicsgo_bulls = pd.read_excel(xlsx_url)
chicsgo_bulls
```

```python
import pandas as pd
xlsx_url = "http://storage.googleapis.com/ds_data_import/fav_nba_teams.xlsx"
boston_celties = pd.read_excel(xlsx_url, sheet_name = 'boston_celtics_2007_2008', skiprows = 6, header = None,
                              names = ['number', 'player', 'pos'], usecols = [0,1,2])
boston_celties
```
